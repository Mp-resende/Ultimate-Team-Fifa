```python
# Código 2 DME
import time
import pyautogui as pgui

# Dicionário com as coordenadas dos slots
COORDENADAS_SLOTS = {
    1: (280, 483),
    2: (745, 523),
    3: (1189, 501),
    4: (1671, 572),
    5: (327, 937),
    6: (772, 931)
}

def clicar_em_slot(slot):
    # Verifica se o slot é válido, caso contrário, usa o padrão (slot 1)
    coordenadas = COORDENADAS_SLOTS.get(slot, COORDENADAS_SLOTS[1])
    pgui.leftClick(x=coordenadas[0], y=coordenadas[1])

def executar_dme(slot):
    clicar_em_slot(slot)
    time.sleep(2)

    # Localiza o template na tela
    posicao = pgui.locateCenterOnScreen("template.png", confidence=0.8)
    if posicao:
        pgui.click(posicao)
        time.sleep(2)
        #Enviar
        pgui.leftClick(x=1382, y=997)
    else:
        print("Imagem não encontrada!")

def main():
    quantidade = int(input("Repetir quantas vezes? "))
    slot = int(input("Qual slot será o DME? (1 a 6) "))

    pgui.PAUSE = 0.8
    pgui.hotkey('alt', 'tab')

    for _ in range(quantidade):
        executar_dme(slot)

if __name__ == "__main__":
    main()
```
